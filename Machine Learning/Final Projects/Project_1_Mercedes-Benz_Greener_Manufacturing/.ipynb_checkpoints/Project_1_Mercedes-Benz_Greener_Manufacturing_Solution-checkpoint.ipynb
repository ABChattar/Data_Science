<h2 style = 'color:Crimson;'>1. If for any column(s), the variance is equal to zero, then you need to remove those variable(s).

```python
X_train = train_df.drop('y', axis=1)
y_train = train_df['y']
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
X_train.dtypes
```

```python
X_train.var()
```

```python
print(X_train.var()['X11'] == 0)
```

```python
for i in X_train:
    if X_train[i].dtypes == 'O':
        continue
    elif (X_train.var()[i] == 0):
        X_train.drop(i, axis= 1, inplace= True)
        test_df.drop(i, axis= 1, inplace= True)
```

<h2 style = 'color:Crimson;'>2. Check for null and unique values for test and train sets.</h2>

<h3 style="color:DarkOrchid;">Check for null values of test and train sets.</h3>

```python
def check_isnull(data):
    for i in data:
        print(f'{i}-->{data[i].isnull().sum()}')
    print("-----------Columns Having Null Values-------------")
    for i in data:
        if(data[i].isnull().sum()> 0):
            print(i)
```

```python
check_isnull(X_train)
```

<h4 style="color:blue;"> Test Dataset</h4>

<h3 style="color:DarkOrchid;">Unique values for test and train sets.</h3>

```python
def unique_values(data):
    for i in data:
        if data[i].dtypes == 'O':
            print(f'For Column {i} --> {data[i].unique()}')
```

<h2 style = 'color:Crimson;'>3. Apply label encoder.</h2>

```python
from sklearn.preprocessing import LabelEncoder
```

```python
col_encoder = LabelEncoder()
```

```python
def col_encoder_fun(data):
    for i in data:
        if data[i].dtypes == 'O':
            data[i] = col_encoder.fit_transform(data[i])
            print(f"Columns That Are Encoded --> {i}")
```

<h4 style="color:blue;"> Train Dataset</h4>

```python
col_encoder_fun(X_train)
```

<h4 style="color:blue;"> Test Dataset</h4>

```python
col_encoder_fun(test_df)
```

<h2 style = 'color:Crimson;'>4. Perform dimensionality reduction.</h2>

<h3 style="color:DarkOrchid;">Normalising the data.</h3> 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(test_df)
```

<h3 style="color:DarkOrchid;">Implementing PCA.</h3> 

```python
from sklearn.decomposition import PCA
```

```python
pca_obj = PCA()
```

```python
X_train_PCA = pca_obj.fit_transform(X_train)
X_test_PCA = pca_obj.fit_transform(X_test)
```

```python
pca_obj.explained_variance_ratio_
```

```python
X_train_PCA.shape
```

<h3 style="color:DarkOrange;">Implementing PCA with 'n_components=170'.</h3> 

```python
pca_obj1 = PCA(n_components=0.98)
```

```python
X_train_pca = pca_obj1.fit_transform(X_train)
```

```python
np.array(pca_obj1.explained_variance_ratio_).sum()*100
```

```python
X_train_pca.shape
```

```python
X_test_pca = pca_obj1.fit_transform(X_test)
```

<h2 style = 'color:Crimson;'>5. Predict your test_df values using XGBoost.</h2>

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train , y_test = train_test_split(X_train_pca, y_train , test_size = 0.2 , random_state = 21)
```

```python
xgb_r = xgb.XGBRegressor(objective ='reg:linear', n_estimators = 7, booster = 'gblinear')
```

```python
xgb_r.fit(X_train, y_train)
```

```python
y_train_pred_xgb = xgb_r.predict(X_train)
```

```python
print("The RMSE value of the model on training dataset is : ")
print(np.sqrt(mean_squared_error(y_train , y_train_pred_xgb)))
```

```python
print("The r2 score of the model on training dataset is : ")
print(r2_score(y_train , y_train_pred_xgb))
```

```python
y_test_pred_xgb = xgb_r.predict(X_test_pca)
```

```python
X_test_Output = pd.DataFrame(test_df["ID"] )
X_test_Output['X_test_output'] = y_test_pred_xgb
```

```python
X_test_Output
```

```python
y_test_pred_xgb = xgb_r.predict(X_test)
```

```python
print("The RMSE value of the model on test dataset is : ")
print(np.sqrt(mean_squared_error(y_test , y_test_pred_xgb)))
```

```python
print("The r2 score of the model on training dataset is : ")
print(r2_score(y_test , y_test_pred_xgb))
```
