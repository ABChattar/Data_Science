## 1. Which movies have maximum views/ratings?

```python
views = []
Movie_names = []
for i in data_df.drop('user_id', axis=1):
    views.append(data_df[i].value_counts().sum())
    Movie_names.append(i)
```

```python
maximum_views = pd.DataFrame()
maximum_views['Movie_names'] = Movie_names
maximum_views['views'] = views
```

```python
maximum_views.iloc[maximum_views['views'].argmax()]
```

### From above we can infer that Movie127 has most views

```python
ratings = []
Movie_names = []
for i in data_df.drop('user_id', axis=1):
    ratings.append(data_df[i].sum())
    Movie_names.append(i)
```

```python
maximum_ratings = pd.DataFrame()
maximum_ratings['Movie_names'] = Movie_names
maximum_ratings['ratings'] = ratings
```

```python
maximum_ratings.iloc[maximum_ratings['ratings'].argmax()]
```

### From above we can infer that Movie127 has most ratings.

## 2. What is the average rating for each movie? Define the top 5 movies with the maximum ratings.

```python
ratings_avg = []
Movie_names = []
for i in data_df.drop('user_id', axis=1):
    ratings_avg.append(data_df[i].mean())
    Movie_names.append(i)
```

- **Recommendation Model: Some of the movies hadnâ€™t been watched and therefore, are not rated by the users. Netflix would like to take this as an opportunity and build a machine learning recommendation algorithm which provides the ratings for each of the users.**

```python
user_ID = []
movie_Name = []
ratings = []
for i in data_df['user_id']:
    temp_df = pd.DataFrame(data_df[data_df['user_id']==i].dropna(axis=1))
    temp_df = temp_df.reset_index()
    for j in temp_df.drop(['user_id', 'index'], axis=1):
        movie_Name.append(j)
        ratings.append(temp_df[j][0])
        user_ID.append(i)   
```

```python
New_data_df = pd.DataFrame()
New_data_df['user_ID'] = user_ID
New_data_df['movie_Name'] = movie_Name
New_data_df['ratings'] = ratings
```

```python
New_data_df.head(5)
```

```python
New_data_df.shape
```

```python
users = pd.DataFrame(New_data_df["user_ID"])
```

## Let's build the association matrix

```python
learningMatrix = New_data_df.pivot_table(index=['user_ID'],columns=['movie_Name'],values='ratings').reset_index(drop=True)
learningMatrix.fillna(0,inplace=True)
learningMatrix.head(20)
```

```python
learningMatrix.shape
```

```python
from sklearn.metrics import pairwise_distances 
```

```python
user_similarity = pairwise_distances(learningMatrix , metric = "cosine")
np.fill_diagonal(user_similarity , 0)
```

```python
similarity_ratings_matrix = pd.DataFrame(user_similarity)
similarity_ratings_matrix.head(15)
```

```python
similarity_ratings_matrix.shape
```

```python
for i in range(0,5):
    list_id = list(New_data_df['user_ID'].sample())
    user_inp = list_id[0]
    try:
        #user_inp=input('Enter the reference user_ID based on which recommendations are to be made: ')
        inp=users[users['user_ID']==user_inp].index.tolist()

        inp=inp[0]

        New_data_df['similarity'] = similarity_ratings_matrix.iloc[inp]
        print("Recommended movies based on your choice of ",user_inp ,": \n", New_data_df.sort_values( ["similarity"], ascending = False )[1:10].drop('user_ID', axis=1))
        print("=============================================================================")
    except:
        print("The user_ID you have entered does not exist in the list, however, below are the top movies recommended in general")
        print(New_data_df.drop('user_ID').head(10))
        print("=============================================================================")
```
